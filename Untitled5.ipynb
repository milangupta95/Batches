```python
import pandas as pd
```

```python
df = pd.read_csv("/content/housing.csv")
```

```python
df.info()
```

```python
median = df.total_bedrooms.median()
```

```python
df['total_bedrooms'] = df.total_bedrooms.fillna(median)
df.info()
```

```python
corM = df.corr()
corM['median_house_value'].sort_values(ascending = False)
```

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(df[['median_house_value','median_income','total_rooms']],figsize=(12,8))
```

```python
df["total_room_per_house"] = df["total_rooms"] / df["households"]
df["total_population_per_house"] = df["population"] / df["households"]
df["total_bedroom_per_house"] = df["total_bedrooms"] / df["households"]
```
